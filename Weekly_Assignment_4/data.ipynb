```python
# create data context
context = gx.get_context()
```

```python
#connect data and make validator
validator = context.sources.pandas_default.read_csv("BankChurners1.csv")
```

```python
#create expectation
validator.expect_column_values_to_be_null("Customer_Age")
validator.expect_column_values_to_be_between("Customer_Age", min_value=20, max_value=80)
validator.expect_table_column_count_to_equal(23)

validator.save_expectation_suite()
```

```python
#run checkpoint
checkpoint = context.add_or_update_checkpoint(
    name="my_quickstart_checkpoint",
    validator=validator,
)
```

```python
checkpoint_result =checkpoint.run()
```

```python
#view validation result
context.view_validation_result(checkpoint_result)
```
